```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
###Merge 12 months of data into a single file
df = pd.read_csv("Pandas-Data-Science-Tasks-master\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data\Sales_April_2019.csv")

files = [file for file in os.listdir('Pandas-Data-Science-Tasks-master\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data')]

all_months_data = pd.DataFrame()
for file in files:
    df = pd.read_csv("Pandas-Data-Science-Tasks-master\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data\{0}".format(file))
    all_months_data = pd.concat([all_months_data, df])

all_months_data.to_csv("all_data.csv", index=False)
```

```python
all_data = pd.read_csv('all_data.csv')
all_data.head()
```

### Cleaning data

```python
#Drop rows of NaN
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()

#Drop the NaN data
all_data = all_data.dropna(how='all')
all_data.head()
```

```python
#Find 'Or' and delete it
all_data = all_data[all_data['Order Date'].str[0:2] !='Or']
all_data.head()
```

```python
#Convert column to the correct type

#Make int
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
#Make float
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])

all_data.head()
```

```python
#Augment dara with additional columns
#Add month column
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

### What was the best month for sale? How much was earned that month?

```python
#Add a sales column
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

```python
result = all_data.groupby('Month').sum()
result
```

```python
months = range(1,13)

plt.bar(months, result['Sales'] )

plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')

plt.show()
```

### What city had the highest number of sales?

```python
all_data.head()
```

```python
#add city column
#use the .apply() method

#grab city
def get_city(address):
    return address.split(',')[1]

#get state
def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x)+ ' ('+get_state(x))+')'
all_data.head()
```

```python
results = all_data.groupby('City').sum()
results
```

```python
cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, results['Sales'] )

plt.xticks(cities, rotation='vertical', size=8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Cities')

plt.show()
```

### What time should we display advertisements to maximise likelihood of customer's buying product?

```python
all_data.head()
```

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data.head()
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours, all_data.groupby('Hour').count())
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of Order')
plt.title('Number of orders by hour')
plt.grid()

plt.show()
```

### What products are most often sold toghter?

```python
all_data.head()
```

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df
```

```python
#group all the duplicate items together

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head()
```

```python
#drop out the duplicate the same order
df = df[['Order ID', 'Grouped']].drop_duplicates()
df
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

for key, value in count.most_common(10):
    print(key, value)
```

### What product sold the most? Why do you think it sold the most?

```python
all_data.head()
```

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']
quantity_ordered
```

```python
products  = [product for product, df in product_group]

plt.bar(products, quantity_ordered)
plt.xticks(products, rotation='vertical', size=8)
plt.ylabel('Quantity Ordered')
plt.xlabel('Products')
plt.title('The quantity of the product that sold the most')
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']
prices
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Order', color='g')
ax2.set_ylabel('Prices ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

plt.show()
```
